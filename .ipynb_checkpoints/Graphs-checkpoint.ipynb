```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df = pd.read_json('FinalDataframe.json')
array_var_obj = ['wardsPlaced - TOPB', 'wardsPlaced - JGB', 'wardsPlaced - MIDB', 'wardsPlaced - ADCB', 'wardsPlaced - SUPPB', 'wardsPlaced - TOPR', 'wardsPlaced - JGR', 'wardsPlaced - MIDR', 'wardsPlaced - ADCR', 'wardsPlaced - SUPPR']
```

```python
df = df[array_var_obj]
```

```python
df.head()
```

```python
warded = []
notWarded = []
ratio = []
for var_obj in array_var_obj:
    warded.append(df[df[var_obj]==1][var_obj].count()/(df[df[var_obj]==0][var_obj].count()+df[df[var_obj]==1][var_obj].count()))
    notWarded.append(df[df[var_obj]==0][var_obj].count()/(df[df[var_obj]==0][var_obj].count()+df[df[var_obj]==1][var_obj].count()))
    ratio.append(df[df[var_obj]==1][var_obj].count()/df[df[var_obj]==0][var_obj].count())
    plt.hist(df[var_obj], bins=2)
    plt.title(var_obj)
    plt.show()
frequencies = pd.DataFrame()
frequencies["Variables"] = array_var_obj
frequencies["Warded"] = warded
frequencies["notWarded"] = notWarded
frequencies["Ratio"] = ratio
```

```python
labels = ["TOP","JG","MID","ADC","SUPP"]
notWarded = [df.groupby("wardsPlaced - TOPB").count().iloc[0,0],df.groupby("wardsPlaced - JGB").count().iloc[0,0],df.groupby("wardsPlaced - MIDB").count().iloc[0,0],df.groupby("wardsPlaced - ADCB").count().iloc[0,0],df.groupby("wardsPlaced - SUPPB").count().iloc[0,0]]
Warded = [df.groupby("wardsPlaced - TOPB").count().iloc[1,0],df.groupby("wardsPlaced - JGB").count().iloc[1,0],df.groupby("wardsPlaced - MIDB").count().iloc[1,0],df.groupby("wardsPlaced - ADCB").count().iloc[1,0],df.groupby("wardsPlaced - SUPPB").count().iloc[1,0]]

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, notWarded, width, label='not warded')
rects2 = ax.bar(x + width/2, Warded, width, label='warded')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frequency')
ax.set_title('Concentration of wards by lanes')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()



fig.tight_layout()

plt.show()
```
